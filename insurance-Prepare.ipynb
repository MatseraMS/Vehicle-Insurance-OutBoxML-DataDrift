```python
# Возраст страхователя на момент последнего продления
df['Age'] = (df['Date_last_renewal'] - df['Date_birth']).dt.days / 365.25
df['Age'] = df['Age'].round(1)

# Стаж вождения
df['Driving_experience'] = (df['Date_last_renewal'] - df['Date_driving_licence']).dt.days / 365.25
df['Driving_experience'] = df['Driving_experience'].round(1)

# Возраст автомобиля
df['Vehicle_age'] = df['Date_last_renewal'].dt.year - df['Year_matriculation']

# До следующего продления (в днях)
df['Days_to_renewal'] = (df['Date_next_renewal'] - df['Date_last_renewal']).dt.days

# Бинарный признак: был ли страховой случай
df['Had_claim'] = (df['N_claims_year'] > 0).astype(int)

# Отношение выплат к премии (индикатор убыточности)
df['Claim_to_premium_ratio'] = df['Cost_claims_year'] / (df['Premium'] + 1)  # +1 чтобы избежать деления на 0

```

```python
def calculate_psi(col: str, current: pd.Series, reference: pd.Series, bins: int = 10, method: str = 'equal_frequency'):
    """
    Вычисляет Population Stability Index (PSI) для одного признака.

    :param col: название признака (для отчёта)
    :param current: текущая выборка (новые данные)
    :param reference: базовая выборка (эталон)
    :param bins: количество бакетов
    :param method: способ биннинга — 'equal_frequency' (дефолт) или 'equal_width'
    :return: словарь с PSI и деталями
    """
    # Удаляем NaN
    current = current.dropna()
    reference = reference.dropna()

    if len(current) == 0 or len(reference) == 0:
        raise ValueError("Одна из выборок пуста после удаления NaN.")

    # Определяем бины по reference
    if method == 'equal_width':
        _, bin_edges = np.histogram(reference, bins=bins)
    elif method == 'equal_frequency':
        bin_edges = np.quantile(reference, np.linspace(0, 1, bins + 1))
        bin_edges[0] = -np.inf
        bin_edges[-1] = np.inf
    else:
        raise ValueError("method должен быть 'equal_width' или 'equal_frequency'")

    # Бинируем обе выборки
    ref_binned = np.digitize(reference, bin_edges)
    curr_binned = np.digitize(current, bin_edges)

    # Подсчитываем доли в каждом бакете
    ref_counts = np.bincount(ref_binned, minlength=bins + 2)[1:-1]  # исключаем -inf и inf
    curr_counts = np.bincount(curr_binned, minlength=bins + 2)[1:-1]

    # Добавляем малое число, чтобы избежать log(0)
    epsilon = 1e-8
    ref_percents = (ref_counts + epsilon) / (len(reference) + epsilon * bins)
    curr_percents = (curr_counts + epsilon) / (len(current) + epsilon * bins)

    # Вычисляем PSI
    diff = curr_percents - ref_percents
    log_ratio = np.log(curr_percents / ref_percents)
    psi_values = diff * log_ratio
    psi = psi_values.sum()

    # Интерпретация
    if psi < 0.1:
        stability = "Стабильное распределение"
    elif psi < 0.25:
        stability = "Умеренное изменение (проверить)"
    else:
        stability = "Сильное изменение (требует вмешательства)"

    result = {
        'psi': round(psi, 4),
        'stability': stability,
        'bins': bins,
        'method': method,
        'bin_edges': [float('inf') if x == np.inf else float('-inf') if x == -np.inf else round(x, 4) for x in bin_edges[1:-1]],
        'reference_distribution': [round(p, 4) for p in ref_percents],
        'current_distribution': [round(p, 4) for p in curr_percents],
        'per_bucket_psi': [round(p, 4) for p in psi_values]
    }

    key = f"psi_{col}"
    return result
```
